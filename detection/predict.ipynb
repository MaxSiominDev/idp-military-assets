```python
from pathlib import Path
from urllib.request import urlretrieve

from ultralytics import YOLO

WEIGHTS_URL  = 'https://storage.yandexcloud.net/idp-web/second/cv/detection/yolov8s_1/best.pt'
WEIGHTS_PATH = Path('/content/best.pt')
SCORE_THRESH = 0.5

urlretrieve(WEIGHTS_URL, WEIGHTS_PATH)
model = YOLO(str(WEIGHTS_PATH))
print('model loaded')
```

```python
import matplotlib.pyplot as plt

src = input('URL: ')

img_path = Path('/content/predict_input.jpg')
urlretrieve(src, img_path)

results = model(str(img_path), conf=SCORE_THRESH)[0]
annotated = results.plot()

plt.figure(figsize=(10, 8))
plt.imshow(annotated[..., ::-1])
plt.axis('off')
plt.title(f'{img_path.name}  detections: {len(results.boxes)}', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```
